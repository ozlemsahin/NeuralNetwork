### MIN MAX SKALIERUNG

```python
#Imports
import random
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
#Create Dataset
#zuf√§llig generierten Datenpunkte aus dem Bereich [100,300]x[0,2]
Y_raw = np.ones(100)
X1_raw = np.ones(200)
X2_raw = np.ones(200)

def randomForX1(X1_raw):
    for i in range(0,200):
        X1_raw[i] = random.uniform(100, 300)
    return

def randomForX2(X2_raw):
    for i in range(0,200):
        X2_raw[i] = random.uniform(0, 2)
    return

randomForX1(X1_raw)
randomForX2(X2_raw)

plt.scatter(X1_raw, X2_raw)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
#Reshape
X1 = X1_raw.reshape(1,X1_raw.shape[0])
X2 = X2_raw.reshape(1,X2_raw.shape[0])
print(X1.shape)
print(X2.shape)
```

```python
#Find min max
x1_min = np.min(X1)
x1_max = np.max(X1)
x2_min = np.min(X2)
x2_max = np.max(X2)
```

```python
#min max algorithm
def minmax(X, x_min , x_max):
    x_mm = np.ones(200)
    x_mm = x_mm.reshape(1,200)
    for i in range(0,200):
        x_mm[0,i] = (X[0,i] - x_min) / (x_max - x_min)
        
    return x_mm


x1_mm = minmax(X1, x1_min, x1_max)
x2_mm = minmax(X2, x2_min, x2_max)
```

```python
plt.scatter(x1_mm, x2_mm)
plt.xlabel("x1 - mm")
plt.ylabel("x2 - mm")
plt.show()
```
